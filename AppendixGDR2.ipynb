```python
import numpy as np
```

```python
import galpy as gp
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
from galpy.potential import MWPotential2014
```

```python
import astropy.units as u 
```

```python
Gaia=pd.read_csv("C:/Users/acer/Downloads/GaiaSource_1584380076484244352_2200921635402776448.csv.gz")
```

```python
s_id=Gaia.source_id
rad_v=Gaia.radial_velocity
rad_v_err=Gaia.radial_velocity_error
dec=Gaia.dec
dec_err=Gaia.dec_error
par=Gaia.parallax
par_err=Gaia.parallax_error
pmra=Gaia.pmra
pmra_err=Gaia.pmra_error
pmdec=Gaia.pmdec
pmdec_err=Gaia.pmdec_error
ra=Gaia.ra
ra_err=Gaia.ra_error
d=1/par
teff=Gaia.teff_val
lum=Gaia.lum_val
data={'source_id': s_id, 'radial_velocity(km/s)': rad_v, 'radial_velocity_error': rad_v_err, 'dec(deg)':dec, 'dec_error':dec_err, 'ra(deg)':ra,'ra_error':ra_err, 'parallax(mas)': par, 'parallax_error':par_err,'pmra(mas/year)': pmra, 'pmra_error':pmra_err, 'pmdec(mas/year)':pmdec, "pmdec_error":pmdec_err, 'distance(pc)':d, "teff_val(K)":teff, "lum_val(Solar Luminosity)":lum}
Gaia=pd.DataFrame(data=data, columns=['source_id', 'radial_velocity(km/s)', 'radial_velocity_error', 'dec(deg)', 'dec_error', 'ra(deg)', 'ra_error', 'parallax(mas)', 'parallax_error', 'pmra(mas/year)', 'pmra_error', 'pmdec(mas/year)', 'pmdec_error','distance(pc)', 'teff_val(K)', 'lum_val(Solar Luminosity)'])
```

```python

stars = Gaia.iloc[:10000]
teff = stars['teff_val(K)']
lum = stars['lum_val(Solar Luminosity)']
plt.figure(figsize=(5,5))
plt.scatter(teff, lum, s=1, c='k')
plt.yscale('log') 
plt.gca().invert_xaxis()   
plt.xlabel("Effective Temperature (K)")
plt.ylabel("Luminosity (L)")
plt.title("Hertzsprung-Russell Diagram")
plt.savefig("HR_diagram.png", dpi=300) 
plt.show()
```

```python
from galpy.orbit import Orbit
star = Gaia.iloc[100]
o = Orbit(vxvv=[star['ra(deg)'], star['dec(deg)'],
                star['pmra(mas/year)'], star['pmdec(mas/year)'],
                star['parallax(mas)'], star['radial_velocity(km/s)']],
          radec=True)
ts = np.linspace(0, 5, 1001)*u.Gyr
o.integrate(ts, MWPotential2014)
o.plot(xrange=[0, 10], yrange=[-10, 10])
plt.title("Orbit integration Star 1")
plt.savefig("Orbit integration Star 1.png", dpi=300)
plt.show()
```
